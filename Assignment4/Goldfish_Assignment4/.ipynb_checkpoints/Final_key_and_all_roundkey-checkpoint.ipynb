```python
import operator
import numpy as np

def hex2bin(s): 
    ans = ""
    maphexa = {'0' : "0000",  
          '1' : "0001", 
          '2' : "0010",  
          '3' : "0011", 
          '4' : "0100", 
          '5' : "0101",  
          '6' : "0110", 
          '7' : "0111",  
          '8' : "1000", 
          '9' : "1001",  
          'A' : "1010", 
          'B' : "1011",  
          'C' : "1100", 
          'D' : "1101",  
          'E' : "1110", 
          'F' : "1111" } 
    for i in range(0,len(s)): 
        ans+=maphexa[s[i]] 
    return ans
      

def bin2hex(s): 
    hex = "" 
    maphexa = {"0000" : '0',  
          "0001" : '1', 
          "0010" : '2',  
          "0011" : '3', 
          "0100" : '4', 
          "0101" : '5',  
          "0110" : '6', 
          "0111" : '7',  
          "1000" : '8', 
          "1001" : '9',  
          "1010" : 'A', 
          "1011" : 'B',  
          "1100" : 'C', 
          "1101" : 'D',  
          "1110" : 'E', 
          "1111" : 'F' } 
    for i in range(0,len(s),4): 
        ind="" 
        ind= ind + s[i] 
        ind= ind+ s[i + 1]  
        ind = ind + s[i + 2]  
        ind = ind + s[i + 3]  
        hex+=maphexa[ind]
          
    return hex
  

def bin2dec(bin):  
    decimal, i, n = 0, 0, 0    
    binary1 = bin
    while(bin != 0):  
        dec = bin % 10
        decimal = decimal + dec * pow(2, i)  
        bin = bin//10
        i= i+1
    return decimal 
  

def dec2bin(num):  
    bin1 = bin(num).replace("0b", "") 
    if(len(bin1)%4 != 0): 
        div = len(bin1) / 4
        div = int(div) 
        counter =(4 * (div + 1)) - len(bin1)  
        for i in range(0, counter): 
            bin1 = '0' + bin1 
    return bin1 
  

def permute(k,arr,n): 
    per = "" 
    for i in range(0, n): 
        per=per+k[arr[i]-1] 
    return per 
  

def shift_left(k,shifts):
    s = ""
    for i in range(shifts):
        n2 = len(k)
        for j in range(1,n2): 
            s+= k[j] 
        s+= k[0]
        k = s
        s = ""  
    return k  
  

def xor(a, b): 
    ans = ""
    for i in range(0,len(a)): 
        if a[i] == b[i]: 
            ans+="0"
        else: 
            ans+="1"
    return ans 


pc1=[                  
  57, 49, 41, 33, 25, 17, 9,
  1, 58, 50, 42, 34, 26, 18,
  10,  2, 59, 51, 43, 35, 27,
  19, 11,  3, 60, 52, 44, 36,
  63, 55, 47, 39, 31, 23, 15, 
  7, 62, 54, 46, 38, 30, 22, 
  14,  6, 61, 53, 45, 37, 29, 
  21, 13,  5, 28, 20, 12,  4
]

pc2 = [                 
  14, 17, 11, 24,  1, 5, 
  3, 28 ,15,  6, 21, 10, 
  23, 19, 12,  4, 26, 8, 
  16,  7, 27, 20, 13, 2, 
  41, 52, 31, 37, 47, 55, 
  30, 40, 51, 45, 33, 48, 
  44, 49, 39, 56, 34, 53, 
  46, 42, 50, 36, 29, 32
]

INV_P = [9, 17, 23, 31,
	      13, 28,  2, 18,
	      24, 16, 30,  6,
	      26, 20, 10,  1,
	       8, 14, 25,  3,
	       4, 29, 11, 19,
	      32, 12, 22,  7,
	       5, 27, 15, 21
         ]

initial_perm = [58, 50, 42, 34, 26, 18, 10, 2,  
                60, 52, 44, 36, 28, 20, 12, 4,  
                62, 54, 46, 38, 30, 22, 14, 6,  
                64, 56, 48, 40, 32, 24, 16, 8,  
                57, 49, 41, 33, 25, 17, 9, 1,  
                59, 51, 43, 35, 27, 19, 11, 3,  
                61, 53, 45, 37, 29, 21, 13, 5,  
                63, 55, 47, 39, 31, 23, 15, 7]  
  
# Expansion E-box Table 
exp_d = [32, 1 , 2 , 3 , 4 , 5 , 4 , 5,  
         6 , 7 , 8 , 9 , 8 , 9 , 10, 11,  
         12, 13, 12, 13, 14, 15, 16, 17,  
         16, 17, 18, 19, 20, 21, 20, 21,  
         22, 23, 24, 25, 24, 25, 26, 27,  
         28, 29, 28, 29, 30, 31, 32, 1 ] 
  
# Straight Permutaion Table 
per = [ 16,  7, 20, 21, 
        29, 12, 28, 17,  
         1, 15, 23, 26,  
         5, 18, 31, 10,  
         2,  8, 24, 14,  
        32, 27,  3,  9,  
        19, 13, 30,  6,  
        22, 11,  4, 25 ] 
  
# S-box Table 
sbox = [[[14, 4, 13, 1, 2, 15, 11, 8, 3 , 10, 6, 12, 5, 9, 0, 7],
    [0, 15, 7, 4, 14, 2, 13, 1, 10, 6, 12, 11, 9, 5, 3, 8],
    [4, 1 , 14, 8, 13, 6, 2, 11, 15, 12, 9, 7,3, 10, 5, 0],
    [15, 12, 8,2,4, 9, 1,7 , 5, 11, 3, 14, 10, 0, 6, 13]],

    [[15, 1, 8, 14, 6, 11, 3, 4, 9, 7, 2, 13, 12, 0,5, 10],
    [3, 13, 4, 7, 15, 2, 8, 14, 12, 0, 1, 10, 6, 9, 11, 5],
    [0, 14, 7, 11, 10, 4, 13, 1, 5, 8,12, 6, 9, 3, 2, 15],
    [13, 8, 10, 1, 3, 15, 4, 2,11,6, 7, 12, 0,5, 14, 9]],

    [[10, 0, 9,14,6,3,15,5, 1, 13, 12, 7, 11, 4,2,8],
    [13, 7, 0, 9, 3, 4, 6, 10, 2, 8, 5, 14, 12, 11, 15, 1],
    [13, 6, 4, 9, 8, 15, 3, 0, 11, 1, 2, 12,5, 10, 14, 7],
    [1, 10, 13, 0, 6, 9, 8, 7, 4, 15, 14, 3, 11, 5, 2, 12]],

    [[7, 13, 14, 3, 0, 6, 9, 10, 1, 2, 8, 5, 11, 12, 4, 15],
    [13, 8, 11, 5, 6, 15, 0, 3, 4, 7, 2, 12, 1, 10, 14, 9],
    [10, 6, 9, 0, 12, 11, 7, 13, 15, 1 , 3, 14, 5, 2, 8, 4],
    [3, 15, 0, 6, 10, 1, 13, 8, 9, 4, 5, 11, 12, 7, 2, 14]],

    [[2, 12, 4, 1, 7, 10, 11, 6, 8, 5, 3, 15, 13, 0, 14, 9],
    [14, 11,2, 12, 4, 7, 13, 1, 5, 0, 15, 10, 3, 9, 8, 6],
    [4, 2, 1, 11, 10, 13, 7, 8, 15, 9, 12, 5, 6, 3, 0, 14],
    [11, 8, 12, 7, 1, 14, 2, 13, 6, 15, 0, 9, 10, 4, 5, 3]],

    [[12, 1, 10, 15, 9, 2, 6,8, 0, 13, 3, 4, 14, 7, 5, 11],
    [10, 15, 4, 2, 7, 12, 9, 5, 6, 1, 13, 14, 0, 11, 3, 8],
    [9, 14, 15, 5, 2,8, 12, 3, 7, 0, 4, 10, 1, 13, 11, 6],
    [4, 3, 2, 12, 9, 5, 15, 10, 11, 14, 1, 7, 6, 0, 8, 13]],

    [[4, 11, 2, 14, 15, 0, 8, 13, 3, 12, 9, 7, 5, 10, 6, 1],
    [13, 0, 11, 7, 4, 9, 1, 10, 14, 3, 5, 12, 2, 15, 8, 6],
    [1, 4, 11, 13, 12, 3, 7, 14, 10, 15, 6, 8, 0, 5, 9, 2],
    [6, 11, 13, 8, 1, 4, 10, 7, 9, 5, 0, 15, 14, 2, 3, 12]],

    [[13, 2, 8, 4, 6, 15, 11, 1, 10, 9, 3, 14, 5, 0, 12,7],
    [1, 15, 13, 8, 10, 3, 7, 4, 12, 5, 6, 11, 0, 14, 9, 2],
    [7, 11, 4, 1, 9, 12, 14, 2, 0, 6, 10, 13, 15, 3, 5, 8],
    [2, 1, 14, 7, 4, 10, 8, 13, 15, 12, 9, 0, 3, 5, 6, 11]]] 
    

final_perm = [ 40, 8, 48, 16, 56, 24, 64, 32,  
               39, 7, 47, 15, 55, 23, 63, 31,  
               38, 6, 46, 14, 54, 22, 62, 30,  
               37, 5, 45, 13, 53, 21, 61, 29,  
               36, 4, 44, 12, 52, 20, 60, 28,  
               35, 3, 43, 11, 51, 19, 59, 27,  
               34, 2, 42, 10, 50, 18, 58, 26,  
               33, 1, 41, 9, 49, 17, 57, 25 ] 

d2s_mapping={'d': '0000',
             'e': '0001',
             'f': '0010',
             'g': '0011',
             'h': '0100',
             'i': '0101',
             'j': '0110',
             'k': '0111',
             'l': '1000',
             'm': '1001',
             'n': '1010',
             'o': '1011',
             'p': '1100',
             'q': '1101',
             'r': '1110',
             's': '1111'}
     

# --parity bit drop table 
keyp = [57, 49, 41, 33, 25, 17, 9,  
        1, 58, 50, 42, 34, 26, 18,  
        10, 2, 59, 51, 43, 35, 27,  
        19, 11, 3, 60, 52, 44, 36,  
        63, 55, 47, 39, 31, 23, 15,  
        7, 62, 54, 46, 38, 30, 22,  
        14, 6, 61, 53, 45, 37, 29,  
        21, 13, 5, 28, 20, 12, 4 ] 


RFP = [57, 49, 41, 33, 25, 17, 9,  1,
       59, 51, 43, 35, 27, 19, 11, 3,
       61, 53, 45, 37, 29, 21, 13, 5,
       63, 55, 47, 39, 31, 23, 15, 7,
       58, 50, 42, 34, 26, 18, 10, 2,
       60, 52, 44, 36, 28, 20, 12, 4,
       62, 54, 46, 38, 30, 22, 14, 6,
       64, 56, 48, 40, 32, 24, 16, 8]

#shift table 
shift_table = [1, 1, 2, 2,  
                2, 2, 2, 2,  
                1, 2, 2, 2,  
                2, 2, 2, 1 ] 
  
# Key- Compression Table : Compression of key from 56 bits to 48 bits 
key_comp = [14, 17, 11, 24, 1, 5,  
            3, 28, 15, 6, 21, 10,  
            23, 19, 12, 4, 26, 8,  
            16, 7, 27, 20, 13, 2,  
            41, 52, 31, 37, 47, 55,  
            30, 40, 51, 45, 33, 48,  
            44, 49, 39, 56, 34, 53,  
            46, 42, 50, 36, 29, 32 ] 
  
    
mapping = {
           'd' : [0,0,0,0],
           'e' : [0,0,0,1],
           'f' : [0,0,1,0],
           'g' : [0,0,1,1],
           'h' : [0,1,0,0],
           'i' : [0,1,0,1],
           'j' : [0,1,1,0],
           'k' : [0,1,1,1],
           'l' : [1,0,0,0],
           'm' : [1,0,0,1],
           'n' : [1,0,1,0],
           'o' : [1,0,1,1],
           'p' : [1,1,0,0],
           'q' : [1,1,0,1],
           'r' : [1,1,1,0],
           's' : [1,1,1,1]
           }
```

```python
def generate_round_keys(k, ro):
    l=k[0:28]     
    r=k[28:56]  
    b=[]
    h=[]
    for i in range(0,ro): 
        #shift bits by n using shift table defined above
        l = shift_left(l, shift_table[i]) 
        r = shift_left(r, shift_table[i]) 
        # merging l and r
        mer=l+r 
        # compressing key from 56 to 48 bits using key compression table from above
        b+=[permute(mer,key_comp,48)] 
        h+=[bin2hex(permute(mer,key_comp,48))]
    return b

def encrypt(pt, kb,ro):
    pt = permute(pt, initial_perm, 64)
    l = pt[0:32] 
    r= pt[32:64] 
    for i in range(0,ro):         
        exp = permute(r, exp_d, 48)   
        r_xor = xor(exp,kb[i])   
        sbox_st= "" 
        for j in range(0,8): 
            row=bin2dec(int(r_xor[j*6] + r_xor[j*6+5])) 
            col=bin2dec(int(r_xor[j*6+1] + r_xor[j*6+2] + r_xor[j*6+3] + r_xor[j*6+4])) 
            value=sbox[j][row][col] 
            sbox_st=sbox_st+dec2bin(value)   
        sbox_st=permute(sbox_st, per, 32)
        res=xor(l, sbox_st)
        l=res
        if(i!=5): 
            temp=l
            l=r
            r=temp
    out=l+r
    ans=permute(out,final_perm,64) 
    return ans

examp = ""
for i in "dddddddddddddddd":
    examp += d2s_mapping[i]
bin_ct = ""
for i in "fgslshkgfkqkiljf":
    bin_ct += d2s_mapping[i]
    
for k in all_possible_keys:
    k_val = generate_round_keys(k,6)
    if(encrypt(examp,k_val,6) == bin_ct):
        
        print("The key is: " + k)
        break
```
