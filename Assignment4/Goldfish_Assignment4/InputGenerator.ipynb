```python
import pandas as pd
import os
import numpy as np
import random
```

```python
letters_to_bits = {
           'd' : '0000',
           'e' : '0001',
           'f' : '0010',
           'g' : '0011',
           'h' : '0100',
           'i' : '0101',
           'j' : '0110',
           'k' : '0111',
           'l' : '1000',
           'm' : '1001',
           'n' : '1010',
           'o' : '1011',
           'p' : '1100',
           'q' : '1101',
           'r' : '1110',
           's' : '1111'
           }

bits_to_letter = {
 '0000': 'd',
 '0001': 'e',
 '0010': 'f',
 '0011': 'g',
 '0100': 'h',
 '0101': 'i',
 '0110': 'j',
 '0111': 'k',
 '1000': 'l',
 '1001': 'm',
 '1010': 'n',
 '1011': 'o',
 '1100': 'p',
 '1101': 'q',
 '1110': 'r',
 '1111': 's'
}
```

```python
def listToString(s):  
    
    s1 = [str(x) for x in s] 
    str1 = "" 
    
    # return string   
    return (str1.join(s1))
```

```python
def ltos(l):
    s = ""
    for i in l:
        s = s+str(i)
    return s
```

```python
# function to calculate pairs of plain text given the xor value
def create_binary_plain(xor):    
    
    XOR = list((bin(xor))[2:].zfill(64))
    XOR = [int(x) for x in XOR]
    binary_plaintexts = []
    
    for i in range(1000):
        tmp=[]
        for j in range(16):
            x= random.choice(['d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s'])
            tmp.append(x)
        inp=''
        for y in tmp:
            inp+=letters_to_bits[y]
        inp = [int(inp[j]) for j in range(len(inp))]
        binary_plaintexts.append(inp)
        binary_plaintexts.append(list(np.bitwise_xor(inp,XOR)))
    
    return binary_plaintexts
```

```python
def convert_binary_to_letter(binaryplain):
    inputs = []

    for i in range(len(binaryplain)):
        input=""
        for j in range(0,64,4):
            temp1 = binaryplain[i][j:j+4]
            temp2 = ltos(temp1)
            input+=bits_to_letter[temp2]
            
        inputs+=[input]
        
    return inputs
```

```python
inputs = create_binary_plain(0x0000801000004000)
inputs = convert_binary_to_letter(inputs)
inputs
```

```python
newfile = open("dummyplain1.txt","w")
for textline in inputs:
    newfile.write(textline+"\n")
newfile.close()
```

```python
inputs = create_binary_plain(0x0000080100100000)
inputs = convert_binary_to_letter(inputs)
inputs
```

```python
newfile = open("dummyplain2.txt","w")
for textline in inputs:
    newfile.write(textline+"\n")
newfile.close()
```
